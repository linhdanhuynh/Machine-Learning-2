```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse
sns.set_theme()
color_list = ['#1f77b4','#ff7f0e','#2ca02c','#d62728','#9467bd','#8c564b','#e377c2','#7f7f7f','#bcbd22','#17becf']

from sklearn import datasets
from sklearn.mixture import GaussianMixture
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
def plot_gmm(gmm, X, ax):
    ax = ax
    labels = gmm.fit(X).predict(X)
    sns.scatterplot(X[:, 0], X[:, 1], hue=labels, ax= ax, s=40, palette='dark:salmon_r', legend= False, alpha=0.5)
 
def plot_score(X):
    n_components = np.arange(2, 21)
    models = [GaussianMixture(n, covariance_type='full', random_state=0).fit(X)
          for n in n_components]
    fig, ax = plt.subplots( 1,2, figsize=(20, 4))
    ax[0].plot(n_components, [m.bic(X) for m in models], label='BIC')
    ax[0].plot(n_components, [m.aic(X) for m in models], label='AIC')

    ax[0].set_xlabel('n_components')
    ax[0].set_title('Elbow Method')

    ax[1].plot(n_components, [silhouette_score(X,m.predict(X)) for m in models])
    ax[1].set_xlabel('n_components')
    ax[1].set_title('Silhouette Score')

def plot_kmeans(kmeans, X ,ax):
    label = kmeans.fit_predict(X)
    sns.scatterplot(X[:, 0] , X[:, 1] , hue= label, ax= ax,palette='dark:salmon_r', legend= False, alpha= 0.5)

    for point in kmeans.cluster_centers_:
        ax.plot(point[0], point[1], '*',markersize=20)
```

```python
K = [2, 3, 4, 5]

X, y = datasets.make_multilabel_classification(n_samples=1000, n_features=2, random_state=0)

fig, ax = plt.subplots( 1,4, figsize=(20, 4))
for i, K in enumerate(K):
    k = GaussianMixture(K)
    plot_gmm(k, X, ax[i])

plt.show()
```

```python
plot_score(X)
```

```python
K = [2, 3, 4, 5]

X, y = datasets.make_blobs(n_samples=1000, n_features=2, random_state=0)

fig, ax = plt.subplots( 1,4, figsize=(20, 4))
for i, K in enumerate(K):
    k = GaussianMixture(K)
    plot_gmm(k, X, ax[i])

plt.show()
```

```python
plot_score(X)
```

```python
K = [2, 3, 4, 5, 6]
X, y = datasets.make_moons(1000, noise=.1, random_state=0)

fig, ax = plt.subplots( 1,5, figsize=(20, 4))
for i, K in enumerate(K):
    k = GaussianMixture(K)
    plot_gmm(k, X, ax[i])

plt.show()
```

```python
plot_score(X)
```

```python
X, y = datasets.make_moons(1500, noise=.1, random_state=0)

fig, ax = plt.subplots( 1,2 ,figsize=(16, 8))

kmeans = KMeans(n_clusters= 4)
plot_kmeans(kmeans, X, ax[0])

k = GaussianMixture(4)
plot_gmm(k, X, ax[1])

plt.legend()
plt.show()
```

```python
X, y = datasets.make_blobs(n_samples=1000, random_state= 170)
transformation = [[0.6, -0.6], [-0.4, 0.8]]
X = np.dot(X, transformation)

fig, ax = plt.subplots( 1,2 ,figsize=(16, 8))

kmeans = KMeans(3)
plot_kmeans(kmeans, X, ax[0])

k = GaussianMixture(3)
plot_gmm(k, X, ax[1])

plt.legend()
plt.show()
```

```python
X, y = datasets.make_blobs(n_samples=1000, cluster_std=[1.0, 2.5, 0.5], random_state=170)

fig, ax = plt.subplots(1, 2,figsize=(16, 8))

kmeans = KMeans(3)
plot_kmeans(kmeans, X, ax[0])

k = GaussianMixture(3)
plot_gmm(k, X, ax[1])

plt.legend()
plt.show()
```

k-Means:  

Advantages	
• Easy	to	implement	
• With	a	large	number	of	variables,	K-Means	may	be	computationally	faster	than hierarchical	clustering	(if	K	is	small).	
• k-Means	may	produce	tighter	clusters	than	hierarchical	clustering	
• An	instance	can	change	cluster	(move	to	another	cluster)	when	the	centroids	are	recomputed.

Disavantages	
• Difficult	to	predict	the	number	of	clusters	(K-Value)	
• Initial	seeds	have	a	strong	impact	on	the	final	results	
• The	order	of	the	data	has	an	impact	on	the	final	results	
• Sensitive	to	scale:	rescaling	your	datasets	(normalizaHon	or	standardization)	will	
completely	change	results.	While	this	itself	is	not	bad,	not	realizing	that	you	have	to	
spend	extra	attention to	scaling	your	data	might	be	bad.	

Gaussian Mixture Model

Advantages:
*   Probabilistic estimates of belonging to each cluster.
*  Does not assume spherical clusters.
*   Handles clusters of differing sizes.
*   Less sensitive to scale.

Disadvantages:
*   Difficult to incorporate categorical features.
*   Assumes a normal distribution for features.
*   Make some assumptions about cluster shape.
*   Somewhat sensitive towards outliers.
*   Somewhat sensitive to initialization conditions.
